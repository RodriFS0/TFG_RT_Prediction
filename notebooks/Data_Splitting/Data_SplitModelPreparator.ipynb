```python
import pandas as pd
from sklearn.model_selection import train_test_split


def split_rt_dataset_stratified_by_id(
    df,
    target_col="rt",
    id_col="id",
    test_size=0.15,
    val_size=0.15,
    random_state=42
):
    """
    Divide el dataset original en train, validation y test
    manteniendo la proporción de cada id en los tres subconjuntos.

    Ejemplo:
    si un id tiene 100 instancias, intentará dejar:
    - 70 en train
    - 15 en validation
    - 15 en test

    Importante:
    - elimina filas sin target válido
    - hace split estratificado por id
    """

    df = df.copy()

    # Convertir target a numérico
    df[target_col] = pd.to_numeric(df[target_col], errors="coerce")

    # Eliminar filas sin target válido
    df = df.dropna(subset=[target_col]).copy()

    # Comprobar que existe la columna id
    if id_col not in df.columns:
        raise ValueError(f"La columna '{id_col}' no existe en el dataframe.")

    # Primer split: train / temp
    train_df, temp_df = train_test_split(
        df,
        test_size=(test_size + val_size),
        random_state=random_state,
        stratify=df[id_col]
    )

    # Tamaño relativo del test dentro de temp
    relative_test_size = test_size / (test_size + val_size)

    # Segundo split: validation / test
    val_df, test_df = train_test_split(
        temp_df,
        test_size=relative_test_size,
        random_state=random_state,
        stratify=temp_df[id_col]
    )

    return train_df, val_df, test_df
```

```python


train_df, val_df, test_df = split_rt_dataset_stratified_by_id(
    df_final,
    target_col="rt",
    id_col="id",
    test_size=0.15,
    val_size=0.15,
    random_state=42
)

# Guardar subconjuntos
train_df.to_csv("train.tsv", sep="\t", index=False)
val_df.to_csv("validation.tsv", sep="\t", index=False)
test_df.to_csv("test.tsv", sep="\t", index=False)

print("Splits guardados correctamente:")
print(f"Train: {train_df.shape}")
print(f"Validation: {val_df.shape}")
print(f"Test: {test_df.shape}")
```

```python
def check_id_proportions(train_df, val_df, test_df, id_col="id"):
    train_counts = train_df[id_col].value_counts().sort_index()
    val_counts = val_df[id_col].value_counts().sort_index()
    test_counts = test_df[id_col].value_counts().sort_index()

    total_counts = train_counts.add(val_counts, fill_value=0).add(test_counts, fill_value=0)

    proportions_df = pd.DataFrame({
        "total": total_counts,
        "train": train_counts,
        "val": val_counts,
        "test": test_counts
    }).fillna(0)

    proportions_df["train_ratio"] = proportions_df["train"] / proportions_df["total"]
    proportions_df["val_ratio"] = proportions_df["val"] / proportions_df["total"]
    proportions_df["test_ratio"] = proportions_df["test"] / proportions_df["total"]

    return proportions_df
```
