```python


import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from google.colab import drive



drive.mount("/content/drive")


PAIRWISE_INPUT_DIR = "/content/drive/MyDrive/TFG_Optuna/Pairwise_Comparisons"

PLOTS_OUTPUT_DIR = "/content/drive/MyDrive/TFG_Optuna/Pairwise_Plots_Final"

os.makedirs(PLOTS_OUTPUT_DIR, exist_ok=True)


PAIRWISE_FILES = {
    "individual_fixed_vs_global_fixed": {
        "path": os.path.join(PAIRWISE_INPUT_DIR, "pairwise_individual_fixed_vs_global_fixed.tsv"),
        "strategy_a": "individual_fixed",
        "strategy_b": "global_fixed",
        "label_a": "Individual fixed",
        "label_b": "Global fixed"
    },

    "individual_fixed_vs_individual_optuna": {
        "path": os.path.join(PAIRWISE_INPUT_DIR, "pairwise_individual_fixed_vs_individual_optuna.tsv"),
        "strategy_a": "individual_fixed",
        "strategy_b": "individual_optuna",
        "label_a": "Individual fixed",
        "label_b": "Individual optimized"
    },

    "global_fixed_vs_global_optuna": {
        "path": os.path.join(PAIRWISE_INPUT_DIR, "pairwise_global_fixed_vs_global_optuna.tsv"),
        "strategy_a": "global_fixed",
        "strategy_b": "global_optuna",
        "label_a": "Global fixed",
        "label_b": "Global optimized"
    },

    "individual_optuna_vs_global_optuna": {
        "path": os.path.join(PAIRWISE_INPUT_DIR, "pairwise_individual_optuna_vs_global_optuna.tsv"),
        "strategy_a": "individual_optuna",
        "strategy_b": "global_optuna",
        "label_a": "Individual optimized",
        "label_b": "Global optimized"
    },
}


METRICS = ["MAE", "RMSE", "R2", "MRE"]

LOWER_IS_BETTER = ["MAE", "RMSE", "MRE"]
HIGHER_IS_BETTER = ["R2"]


# NUMERIC CLEANING

def parse_metric_value(x):
    """
    Robust parser for normal numeric values and old malformed values.

    Normal example:
        0.534

    Old malformed example:
        843.843.129.987.644

    For malformed values, this parser assumes the decimal point was lost.
    """

    if pd.isna(x):
        return np.nan

    if isinstance(x, (int, float, np.integer, np.floating)):
        return float(x)

    s = str(x).strip()

    if s == "":
        return np.nan

    s = s.replace(",", ".")

    try:
        return float(s)
    except Exception:
        pass

    sign = -1 if s.startswith("-") else 1
    s_clean = s.lstrip("+-")

    digits = "".join(ch for ch in s_clean if ch.isdigit())

    if digits == "":
        return np.nan

    if len(digits) == 1:
        return sign * float(digits)

    value = float(digits[0] + "." + digits[1:])

    return sign * value


def metric_better_direction(metric):
    if metric in LOWER_IS_BETTER:
        return "lower"

    if metric in HIGHER_IS_BETTER:
        return "higher"

    raise ValueError(f"Metric direction not defined: {metric}")


def get_improvement_b_over_a(metric, x_a, x_b):
    """
    Positive improvement means strategy B is better than strategy A.

    MAE, RMSE, MRE:
        improvement = strategy_a - strategy_b

    R2:
        improvement = strategy_b - strategy_a
    """

    diff_b_minus_a = x_b - x_a

    if metric_better_direction(metric) == "lower":
        return -diff_b_minus_a

    return diff_b_minus_a


def clean_pairwise_metric_columns(df, strategy_a, strategy_b, metrics):
    df = df.copy()

    if "id" not in df.columns:
        raise ValueError("Missing column: id")

    df["id"] = df["id"].astype(str).str.strip()

    for metric in metrics:
        col_a = f"{metric}_{strategy_a}"
        col_b = f"{metric}_{strategy_b}"

        if col_a not in df.columns:
            raise ValueError(f"Missing column: {col_a}")

        if col_b not in df.columns:
            raise ValueError(f"Missing column: {col_b}")

        df[col_a] = df[col_a].apply(parse_metric_value)
        df[col_b] = df[col_b].apply(parse_metric_value)

        n_missing_a = df[col_a].isna().sum()
        n_missing_b = df[col_b].isna().sum()

        if n_missing_a > 0:
            print(f"Warning: {col_a} has {n_missing_a} missing values after parsing.")

        if n_missing_b > 0:
            print(f"Warning: {col_b} has {n_missing_b} missing values after parsing.")

    return df


def build_improvement_dataframe(df, strategy_a, strategy_b, metrics):
    rows = []

    for metric in metrics:
        col_a = f"{metric}_{strategy_a}"
        col_b = f"{metric}_{strategy_b}"

        tmp = df[["id", col_a, col_b]].copy()
        tmp = tmp.replace([np.inf, -np.inf], np.nan).dropna()

        x_a = tmp[col_a].to_numpy(dtype=float)
        x_b = tmp[col_b].to_numpy(dtype=float)

        improvement = get_improvement_b_over_a(metric, x_a, x_b)
        diff_b_minus_a = x_b - x_a

        for i, row_id in enumerate(tmp["id"].tolist()):
            rows.append({
                "id": row_id,
                "metric": metric,
                "value_a": x_a[i],
                "value_b": x_b[i],
                "diff_b_minus_a": diff_b_minus_a[i],
                "improvement_b_over_a": improvement[i]
            })

    return pd.DataFrame(rows)


def bootstrap_ci_median(values, n_boot=2000, ci=95, seed=42):
    values = np.asarray(values, dtype=float)
    values = values[np.isfinite(values)]

    if len(values) == 0:
        return np.nan, np.nan, np.nan

    rng = np.random.default_rng(seed)

    boot_medians = []

    for _ in range(n_boot):
        sample = rng.choice(values, size=len(values), replace=True)
        boot_medians.append(np.median(sample))

    boot_medians = np.asarray(boot_medians, dtype=float)

    alpha = (100 - ci) / 2

    median_value = float(np.median(values))
    lower = float(np.percentile(boot_medians, alpha))
    upper = float(np.percentile(boot_medians, 100 - alpha))

    return median_value, lower, upper


def save_plot_summary(improvement_df, comparison_name, output_dir):
    rows = []

    for metric in METRICS:
        tmp = improvement_df[improvement_df["metric"] == metric].copy()
        values = tmp["improvement_b_over_a"].to_numpy(dtype=float)
        values = values[np.isfinite(values)]

        if len(values) == 0:
            continue

        median_value, ci_low, ci_high = bootstrap_ci_median(values)

        rows.append({
            "comparison": comparison_name,
            "metric": metric,
            "n_pairs": int(len(values)),
            "mean_improvement_b_over_a": float(np.mean(values)),
            "median_improvement_b_over_a": median_value,
            "bootstrap_median_ci_low": ci_low,
            "bootstrap_median_ci_high": ci_high,
            "strategy_b_better_count": int(np.sum(values > 0)),
            "strategy_a_better_count": int(np.sum(values < 0)),
            "ties_count": int(np.sum(values == 0)),
            "strategy_b_better_percent": float(np.sum(values > 0) / len(values) * 100),
            "strategy_a_better_percent": float(np.sum(values < 0) / len(values) * 100),
            "ties_percent": float(np.sum(values == 0) / len(values) * 100),
        })

    summary_df = pd.DataFrame(rows)

    output_path = os.path.join(
        output_dir,
        f"{comparison_name}_plot_summary.tsv"
    )

    summary_df.to_csv(output_path, sep="\t", index=False)

    print(f"Saved summary: {output_path}")

    return summary_df


def save_improvement_long_table(improvement_df, comparison_name, output_dir):
    output_path = os.path.join(
        output_dir,
        f"{comparison_name}_improvement_long.tsv"
    )

    improvement_df.to_csv(output_path, sep="\t", index=False)

    print(f"Saved improvement table: {output_path}")


#PLOT 1
# IMPROVEMENT DISTRIBUTION

def plot_improvement_distribution(
    improvement_df,
    comparison_name,
    label_a,
    label_b,
    output_path
):
    fig, axes = plt.subplots(2, 2, figsize=(12, 8))
    axes = axes.flatten()

    for ax, metric in zip(axes, METRICS):
        tmp = improvement_df[improvement_df["metric"] == metric].copy()
        values = tmp["improvement_b_over_a"].to_numpy(dtype=float)
        values = values[np.isfinite(values)]

        ax.boxplot(
            values,
            vert=False,
            showmeans=True
        )

        ax.axvline(0, linestyle="--", linewidth=1)

        ax.set_title(metric)
        ax.set_xlabel(f"Improvement of {label_b} over {label_a}")
        ax.set_yticks([])

        ax.grid(axis="x", alpha=0.3)

        if metric in LOWER_IS_BETTER:
            subtitle = "Positive = lower error with strategy B"
        else:
            subtitle = "Positive = higher R2 with strategy B"

        ax.text(
            0.5,
            0.95,
            subtitle,
            transform=ax.transAxes,
            ha="center",
            va="top",
            fontsize=9
        )

    fig.suptitle(
        f"{comparison_name}: paired improvement distribution",
        fontsize=14
    )

    plt.tight_layout()
    plt.savefig(output_path, dpi=300, bbox_inches="tight")
    plt.show()

    print(f"Saved: {output_path}")


# PLOT 2
# RANKED IMPROVEMENT BY ID


def plot_ranked_improvement(
    improvement_df,
    comparison_name,
    label_a,
    label_b,
    output_path
):
    fig, axes = plt.subplots(2, 2, figsize=(12, 8))
    axes = axes.flatten()

    for ax, metric in zip(axes, METRICS):
        tmp = improvement_df[improvement_df["metric"] == metric].copy()

        values = tmp["improvement_b_over_a"].to_numpy(dtype=float)
        values = values[np.isfinite(values)]
        values_sorted = np.sort(values)

        x = np.arange(1, len(values_sorted) + 1)

        ax.plot(
            x,
            values_sorted,
            marker="o",
            markersize=2,
            linewidth=1
        )

        ax.axhline(0, linestyle="--", linewidth=1)

        ax.set_title(metric)
        ax.set_xlabel("IDs ordered by improvement")
        ax.set_ylabel(f"Improvement of {label_b} over {label_a}")

        ax.grid(alpha=0.3)

    fig.suptitle(
        f"{comparison_name}: ranked paired improvement",
        fontsize=14
    )

    plt.tight_layout()
    plt.savefig(output_path, dpi=300, bbox_inches="tight")
    plt.show()

    print(f"Saved: {output_path}")


#  PLOT 3
# OVERLAY DISTRIBUTIONS


def plot_overlay_distributions(
    df,
    comparison_name,
    strategy_a,
    strategy_b,
    label_a,
    label_b,
    output_path,
    bins=30,
    percentile_xlim=99
):
    fig, axes = plt.subplots(2, 2, figsize=(12, 8))
    axes = axes.flatten()

    for ax, metric in zip(axes, METRICS):
        col_a = f"{metric}_{strategy_a}"
        col_b = f"{metric}_{strategy_b}"

        tmp = df[[col_a, col_b]].copy()
        tmp = tmp.replace([np.inf, -np.inf], np.nan).dropna()

        values_a = tmp[col_a].to_numpy(dtype=float)
        values_b = tmp[col_b].to_numpy(dtype=float)

        all_values = np.concatenate([values_a, values_b])
        all_values = all_values[np.isfinite(all_values)]

        if len(all_values) == 0:
            continue

        lower = np.percentile(all_values, 100 - percentile_xlim)
        upper = np.percentile(all_values, percentile_xlim)

        ax.hist(
            values_a,
            bins=bins,
            density=True,
            alpha=0.45,
            label=label_a
        )

        ax.hist(
            values_b,
            bins=bins,
            density=True,
            alpha=0.45,
            label=label_b
        )

        ax.axvline(np.median(values_a), linestyle="--", linewidth=2)
        ax.axvline(np.median(values_b), linestyle="--", linewidth=2)

        if lower < upper:
            ax.set_xlim(lower, upper)

        ax.set_title(metric)
        ax.set_xlabel(metric)
        ax.set_ylabel("Density")

        if metric in LOWER_IS_BETTER:
            subtitle = "Lower is better"
        else:
            subtitle = "Higher is better"

        ax.text(
            0.5,
            0.95,
            subtitle,
            transform=ax.transAxes,
            ha="center",
            va="top",
            fontsize=9
        )

        ax.grid(axis="y", alpha=0.3)
        ax.legend()

    fig.suptitle(
        f"{comparison_name}: overlapping metric distributions",
        fontsize=14
    )

    plt.tight_layout()
    plt.savefig(output_path, dpi=300, bbox_inches="tight")
    plt.show()

    print(f"Saved: {output_path}")


#PLOT 4
# PAIRED SLOPE PLOT


def plot_paired_slope(
    df,
    comparison_name,
    strategy_a,
    strategy_b,
    label_a,
    label_b,
    output_path,
    max_lines=250
):
    fig, axes = plt.subplots(2, 2, figsize=(12, 8))
    axes = axes.flatten()

    for ax, metric in zip(axes, METRICS):
        col_a = f"{metric}_{strategy_a}"
        col_b = f"{metric}_{strategy_b}"

        tmp = df[["id", col_a, col_b]].copy()
        tmp = tmp.replace([np.inf, -np.inf], np.nan).dropna()

        if len(tmp) > max_lines:
            tmp_lines = tmp.sample(n=max_lines, random_state=42)
        else:
            tmp_lines = tmp

        for _, row in tmp_lines.iterrows():
            ax.plot(
                [1, 2],
                [row[col_a], row[col_b]],
                alpha=0.18,
                linewidth=0.8
            )

        ax.scatter(
            np.repeat(1, len(tmp_lines)),
            tmp_lines[col_a].to_numpy(dtype=float),
            alpha=0.45,
            s=12
        )

        ax.scatter(
            np.repeat(2, len(tmp_lines)),
            tmp_lines[col_b].to_numpy(dtype=float),
            alpha=0.45,
            s=12
        )

        median_a = tmp[col_a].median()
        median_b = tmp[col_b].median()

        ax.plot(
            [1, 2],
            [median_a, median_b],
            linewidth=3
        )

        ax.set_xticks([1, 2])
        ax.set_xticklabels([label_a, label_b], rotation=10)

        ax.set_title(metric)
        ax.set_ylabel(metric)

        ax.grid(axis="y", alpha=0.3)

    fig.suptitle(
        f"{comparison_name}: paired slope plot by id",
        fontsize=14
    )

    plt.tight_layout()
    plt.savefig(output_path, dpi=300, bbox_inches="tight")
    plt.show()

    print(f"Saved: {output_path}")



#  PLOT 5
# ESTIMATION PLOT


def plot_estimation_median_ci(
    improvement_df,
    comparison_name,
    label_a,
    label_b,
    output_path,
    n_boot=2000,
    ci=95
):
    fig, axes = plt.subplots(2, 2, figsize=(12, 8))
    axes = axes.flatten()

    for ax, metric in zip(axes, METRICS):
        tmp = improvement_df[improvement_df["metric"] == metric].copy()

        values = tmp["improvement_b_over_a"].to_numpy(dtype=float)
        values = values[np.isfinite(values)]

        median_value, ci_low, ci_high = bootstrap_ci_median(
            values,
            n_boot=n_boot,
            ci=ci,
            seed=42
        )

        ax.hist(
            values,
            bins=30,
            alpha=0.65
        )

        ax.axvline(0, linestyle="--", linewidth=1)
        ax.axvline(median_value, linewidth=2)
        ax.axvline(ci_low, linestyle=":", linewidth=2)
        ax.axvline(ci_high, linestyle=":", linewidth=2)

        ax.set_title(metric)
        ax.set_xlabel(f"Improvement of {label_b} over {label_a}")
        ax.set_ylabel("Number of ids")

        ax.grid(axis="y", alpha=0.3)

        ax.text(
            0.5,
            0.95,
            f"Median = {median_value:.4f}\n{ci}% CI = [{ci_low:.4f}, {ci_high:.4f}]",
            transform=ax.transAxes,
            ha="center",
            va="top",
            fontsize=9
        )

    fig.suptitle(
        f"{comparison_name}: estimation of median paired improvement",
        fontsize=14
    )

    plt.tight_layout()
    plt.savefig(output_path, dpi=300, bbox_inches="tight")
    plt.show()

    print(f"Saved: {output_path}")




def plot_one_pairwise_file_final(
    comparison_name,
    pairwise_path,
    strategy_a,
    strategy_b,
    label_a,
    label_b,
    output_dir=PLOTS_OUTPUT_DIR
):
    if not os.path.exists(pairwise_path):
        raise FileNotFoundError(f"File not found: {pairwise_path}")

    print(f"Loading: {pairwise_path}")

    df = pd.read_csv(pairwise_path, sep="\t")

    df = clean_pairwise_metric_columns(
        df=df,
        strategy_a=strategy_a,
        strategy_b=strategy_b,
        metrics=METRICS
    )

    improvement_df = build_improvement_dataframe(
        df=df,
        strategy_a=strategy_a,
        strategy_b=strategy_b,
        metrics=METRICS
    )

    save_improvement_long_table(
        improvement_df=improvement_df,
        comparison_name=comparison_name,
        output_dir=output_dir
    )

    summary_df = save_plot_summary(
        improvement_df=improvement_df,
        comparison_name=comparison_name,
        output_dir=output_dir
    )

    plot_improvement_distribution(
        improvement_df=improvement_df,
        comparison_name=comparison_name,
        label_a=label_a,
        label_b=label_b,
        output_path=os.path.join(
            output_dir,
            f"{comparison_name}_01_improvement_distribution.png"
        )
    )

    plot_ranked_improvement(
        improvement_df=improvement_df,
        comparison_name=comparison_name,
        label_a=label_a,
        label_b=label_b,
        output_path=os.path.join(
            output_dir,
            f"{comparison_name}_02_ranked_improvement.png"
        )
    )

    plot_overlay_distributions(
        df=df,
        comparison_name=comparison_name,
        strategy_a=strategy_a,
        strategy_b=strategy_b,
        label_a=label_a,
        label_b=label_b,
        output_path=os.path.join(
            output_dir,
            f"{comparison_name}_03_overlay_distributions.png"
        )
    )

    plot_paired_slope(
        df=df,
        comparison_name=comparison_name,
        strategy_a=strategy_a,
        strategy_b=strategy_b,
        label_a=label_a,
        label_b=label_b,
        output_path=os.path.join(
            output_dir,
            f"{comparison_name}_04_paired_slope_plot.png"
        )
    )

    plot_estimation_median_ci(
        improvement_df=improvement_df,
        comparison_name=comparison_name,
        label_a=label_a,
        label_b=label_b,
        output_path=os.path.join(
            output_dir,
            f"{comparison_name}_05_estimation_median_ci.png"
        )
    )

    return df, improvement_df, summary_df



#  RUN ALL PAIRWISE FILES


def plot_all_pairwise_files_final(
    pairwise_files,
    output_dir=PLOTS_OUTPUT_DIR
):
    os.makedirs(output_dir, exist_ok=True)

    loaded_data = {}

    for comparison_name, cfg in pairwise_files.items():
        print("=" * 80)
        print(f"Plotting: {comparison_name}")

        df, improvement_df, summary_df = plot_one_pairwise_file_final(
            comparison_name=comparison_name,
            pairwise_path=cfg["path"],
            strategy_a=cfg["strategy_a"],
            strategy_b=cfg["strategy_b"],
            label_a=cfg["label_a"],
            label_b=cfg["label_b"],
            output_dir=output_dir
        )

        loaded_data[comparison_name] = {
            "pairwise_df": df,
            "improvement_df": improvement_df,
            "summary_df": summary_df
        }

    return loaded_data


loaded_pairwise_plot_data = plot_all_pairwise_files_final(
    pairwise_files=PAIRWISE_FILES,
    output_dir=PLOTS_OUTPUT_DIR
)

print("All final plots saved in:")
print(PLOTS_OUTPUT_DIR)
```
